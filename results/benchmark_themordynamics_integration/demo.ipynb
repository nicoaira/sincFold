## Fold a short sequence

```python
!sincFold pred "AACCGGGUCAGGUCCGGAAGGAAGCAGCCCUAA"
```

Adding drawing options

```python
!sincFold pred "AACCGGGUCAGGUCCGGAAGGAAGCAGCCCUAA" --name srp_Bruc.abor._AE017223 --draw --draw-resolution .8

from PIL import Image
Image.open("srp_Bruc.abor._AE017223.png")
```

## Folding a longer sequence

```python
!sincFold pred "CCACGGCGACUAUAUCCCUGGUGUUCACCUCUUCCCAUUCCGAACAGAGUCGUUAAGCCCAGGAGAGCCGAUGGUACUGCUUUAUUGCGGGAGAGUAGGUCGUCGCCGAGU" --name longer_seq --draw
Image.open("longer_seq.png")
```

## Using .fasta files

```python
# creating file with two sequences
!echo -e ">seq1\\nAACCGGGUCAGGUCCGGAAGGAAGCAGCCCUAA" > test.fasta
!echo -e ">seq2\\nGUAGUCGUGGCCGAGUGGUUAAGGCGAUGGACUAGAAAUCCAUUGGGGUCUCCCCGCGCAGGUUCGAAUCCUGCCGACUACGCCA" >> test.fasta
```

```python
!sincFold pred test.fasta
```

Outputs in .ct format

```python
!sincFold pred test.fasta -o test_outputs
```

```python
!ls test_outputs
```

Outputs in .csv format

```python
!sincFold pred test.fasta -o test_outputs.csv
!cat test_outputs.csv
```
